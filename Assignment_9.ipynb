To study and implement the ORB (Oriented FAST and Rotated BRIEF) algorithm for detecting and matching key features between images, enabling fast and efficient feature-based image comparison and recognition.

```python
import cv2
import matplotlib.pyplot as plt

# Load images
img1 = cv2.imread('pic5.jpg', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('pic6.jpg', cv2.IMREAD_GRAYSCALE)

# Initialize ORB detector
orb = cv2.ORB_create()

# Detect keypoints and compute descriptors
keypoints1, descriptors1 = orb.detectAndCompute(img1, None)
keypoints2, descriptors2 = orb.detectAndCompute(img2, None)

# Match descriptors using BFMatcher
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
matches = bf.match(descriptors1, descriptors2)
matches = sorted(matches, key=lambda x: x.distance)

# Draw matches
matched_img = cv2.drawMatches(img1, keypoints1, img2, keypoints2, matches[:50], None, flags=2)

# Show results
plt.figure(figsize=(10,5))
plt.imshow(matched_img)
plt.title("ORB Feature Matching")
plt.axis('off')
plt.show()
```
